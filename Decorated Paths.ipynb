# 24 Decorated Paths

```python
%load_ext tikzmagic
```

```python
%%tikz -l decorations.pathreplacing

\draw[decorate, decoration={brace}] (0,0) -- (3,0);
\draw[decorate, decoration={brace,mirror}] (0,1) -- (3,1);
```

```python
%%tikz -l decorations.pathmorphing

\draw[decorate, decoration={zigzag}] (0,0) -- (3,0);
\draw[decorate, decoration={zigzag,mirror}] (0,1) -- (3,1);
\draw[decorate, decoration={bumps}] (0,2) -- (3,2);
\draw[decorate, decoration={bumps,mirror}] (0,3) -- (3,3);
```
