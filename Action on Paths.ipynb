## color

* `definecolor{name}{rgb}{r숫자, g숫자, b숫자}`: 각 숫자는 0부터 1사이


* `color=colorname!숫자`, `color=colorname!숫자!colorname2`. text, draw, fill 색상. 숫자 0(가장 밝은 색)부터 100(가장 어두운 색)
* `text=`: 노드 글자 색상
* `draw=`: 선 색상
* `fill=`: 내부 색상
* `pattern color=`: 내부 패턴 색상
* `double=`: 두겹선 내부 색상

```python
%%tikz

\definecolor{orange}{rgb}{1,0.5,0};
\path[color=green, text=orange, draw=blue!90, fill=blue!10, double] (0,0) rectangle (1,1) node {text};
```

## draw action


* `line width=0.4pt`
  * `ultra thin`: 0.1pt
  * `very thin`: 0.2pt
  * `thin`: 0.4pt
  * `semitick`: 0.6pt
  * `thick`: 0.8pt
  * `very thick`: 1.2pt
  * `ultra thick`: 1.6pt

* `dash pattern= on 2pt off 3pt`
* `dash phase=0pt`
  * `solid`
  * `dotted`, `loosely dotted`, `densely dotted`
  * `dashed`, `loosely dashed`, `densely dashed`
  * `dash dot`, `loosely dash dot`, `densely dash dot`
  * `dash dot dot`, `loosely dash dot dot`, `densely dash dot dot`
  

## fill action

* `--`, `circle`, `rectangle` operation에 의해 생성된 모든 공간에 색칠
* 닫힌 경로가 아니더라도 색칠


* `pattern` 라이브러리

  * `horizontal lines`
  * `vertical lines`
  * `north east lines`
  * `north west lines`
  * `grid`
  * `crosshatch`
  * `dots`
  * `crosshatch dots`
  * `fivepointed stars`
  * `sixpointed stars`
  * `bricks`
  * `checkerboard`

```python
%%tikz -l patterns

\draw[pattern color=green!40!black,pattern=checkerboard] (0,0) -- (5,5) -- (5,0);
```
