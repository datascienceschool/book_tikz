# 56 Math

`\tikzmath {}`

* 빈 줄이 있으면 안됨

```python
%%tikz  -l math

\tikzmath {
    int \i;
    \i = 10;
    real \x;
    \x = 3.14;
    coordinate \c;
    \c = (0,0)+(1,1);
}

\node at (\c) {{\tt\string\i} is \i, {\tt\string\x} is \x};
```

* 변수는 숫자 혹은 문자 인덱스를 가질 수 있음

```python
%%tikz -l math

\tikzmath {
    int \x;
    \x1 = 3+4;
    \x2 = 30+40;
    \x3 = 300+400;
}

\foreach \i in {1,2,3} {
    \node at (\i, 0) {\x\i};
}
```

```python
%%tikz -l math

\tikzmath {
    int \c;
    \c{air} = 340;
    \c{water} = 1435;
    \c{steel} = 6100;
}

\foreach \i/\m in {0/air, 1/water, 2/steel} {
    \node at (\i,0) {\c{\m}};
}
```
