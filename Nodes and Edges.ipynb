# 17 Nodes and Edges

$$
\hbox{\path} \ldots \hbox{node} 
\overbrace{
    \, \hbox{foreach} \, \underbrace{
        [\hbox{option}] \, (\hbox{name}) \, \hbox{at} \, (\hbox{coordinate})
    }_{\hbox{순서 상관없음}}
}^{\hbox{필수 아님}}
{\hbox{{text}}}
$$

    * draw=
    * fill=
    
  * shape options
  
    * rectangle

    * coordinate
    
  * depth

  * style
  
    * every node=
    * every rectangle node=
    * every circle node=
    * every coordinate node=
    
  * size
  
    * inner sep=

    * inner yset=
    * outer sep=: `auto`
    * outer xsep=
    * outer ysep=
    * minimum size=
    * minimum height=
    * minimum width=
  
  * rotate
  
    * shape border rotate=
    
  * text
  
    * node font=
    * font=

    * text height=
    * text depth=
    
  * position
  
    * anchor=: `center`, `base`, `north`, `south`, `east`, `west`, ...
    * above=
    * below=
    * left=
    * right=
    * pos=
    * sloped

* color 옵션은 text, draw, fill 모두에 적용된다. 각각의 색상 옵션과 color 옵션을 동시에 사용할 때는 순서에 주의한다. color 옵션이 가장 앞에 나와야 한다.

```python
%%tikz -p kotex

\node[black,draw,text=red] {$\alpha + \hbox{길이가 긴 english와 한글}$};
```

* path에 대한 draw, fill 명령은 노드에는 적용되지 않는다. 각각의 노드에 대해 별도로 draw, fill 명령을 주어야 한다. 

```python
%%tikz 

\draw (0,0) -- (1,1) node {text};
\draw (0,0) -- (1,0) node[draw] {text};
\fill (-1,0) -- (-1,1) -- (-2,1) node {text};
\fill (-1,0) -- (0,1) -- (0,0) node[fill=blue] {text};
```

```python
%%tikz 

\node foreach \x in {1,...,4} foreach \y in {1,...,4} [draw] (\x/\y) at (\x, \y) {\x,\y};

\draw (1/2) -- (2/3);
```

* pos 옵션은 경로상의 마지막 **"라인"**상에 노드를 위치시킨다.

```python
%%tikz

\draw (0,0) grid (10, 1);
\draw (0,0) -- (5,0) (6,0)    (10, 0) node[pos=0.2] {0.2} node[pos=0.5] {0.5} node[pos=0.8] {0.8};
\draw (0,1) -- (5,1) (6,1) -- (10, 1) node[pos=0.2] {0.2} node[pos=0.5] {0.5} node[pos=0.8] {0.8};
```

```python
%%tikz 

\node[draw] (a) at (0,0) {a};
\node[draw] (b) at (1,1) {b};
\draw (a) -| (b) node[pos=0.8, right] {c} ;
```
