# 25 Transformation

* `x`, `y`, `z`
* `shift={()}`, `xshift=dim`, `yshift=dim`
* `rotate`, `rotate around x`, `rotate around y`, `rotate around z`
* `scale`, `xscale`, `yscale`


```python
%load_ext tikzmagic
```

```python
%%tikz

\draw[help lines] (0,-5) grid (25,5);
\fill (0,0) circle (3mm);

\draw[shift={(1,2)}, xscale=1, rotate=45] grid (2,2);
```

```python
%%tikz

\def\mygrid#1#2{%
    \foreach \x in {0,...,#1} {
        \foreach \y in {0,...,#2} {
            \filldraw[fill=gray] (\x,\y) -- ++(1,0) -- ++(0,1) -- ++(-1,0) -- cycle;
        }
    } 
}

\begin{scope}[x={(0.95cm,-0.1cm)},z={(-0.8cm,-0.2cm)}]
    \foreach \i in {0,5,10,15,20,25} {  
        \begin{scope}[shift={(\i,0)}, rotate around y=-90]
            \mygrid {10} {10}
        \end{scope}
    }
\end{scope}
```
