```python
%load_ext tikzmagic
```

`\matrix (name) [options] { };`

* matrix of nodes
* matrix of math nodes


노드별 옵션은 `|[]|` 안에

```python
%%tikz -l matrix

\matrix (magic) [matrix of nodes] {
    |[draw=red,circle]|8 & 1 & 6 \\
    3 & 5 & |[draw=red,circle]|7 \\
    4 & 9 & 2 \\
};

\draw[red,->] (magic-1-1) |- (magic-2-3);
```

모든 노드에 대한 옵션은 `nodes={}`

```python
%%tikz -l matrix

\matrix (magic) [
    matrix of math nodes, 
    nodes={draw=red,circle,minimum size=30pt},
    left delimiter=(,right delimiter=\},
    above delimiter=\|,below delimiter=\rmoustache                
] {
    a_8 & \phantom{a_2} & a_6 \\
    a_3 & a_5 & a_7 \\
    a_4 & a_9 & a_2 \\
};
```
