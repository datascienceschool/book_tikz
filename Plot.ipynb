# 22 Plot

`\path plot (\x,{exp(-\x*\x)})`

* options

  * `domain=`: default `-5:5`
  * `samples=`: default `25`

```python
%%tikz 

\draw plot[smooth] (\x,{5*exp(-0.2*\x*\x)});
```

```python
%%tikz

\draw[help lines] (0,0) grid (10,5);

\tikzset{normalplot/.pic={\draw plot[rotate=-90, domain=-1:1] (\x,{2*exp(-4*\x*\x)});}}

\pic at (7,4) {normalplot};
\pic at (1,1) {normalplot};
```
